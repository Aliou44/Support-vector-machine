```python
from sklearn.preprocessing import OneHotEncoder

# Selectionne les colonnes catégorielles
x_cat = df.select_dtypes('object').drop(columns=["Individual ID", "Comments", "Species"], axis=1)

# Remplace les NA par le mode de chaque colonne
x_cat.fillna(x_cat.mode().iloc[0], inplace=True)

# Initialise le One Hot Encoding
ohe = OneHotEncoder(handle_unknown = 'ignore')

# Détection des classes
ohe.fit(x_cat)

# Transformation des colonnes par One Hot
x_cat_ohe = ohe.transform(x_cat)
x_cat_ohe
```

```python
x_cat_ohe = x_cat_ohe.toarray()
x_cat_ohe
```

```python
x_cat_ohe.shape
```

```python
ohe.feature_names_in_
```

```python
ohe.get_feature_names_out()
```

```python
x_cat_ohe = pd.DataFrame(x_cat_ohe, columns=ohe.get_feature_names_out())
x_cat_ohe
```

```python
from sklearn.preprocessing import StandardScaler
x_num = df.select_dtypes(['int', 'float']).drop('Sample Number', axis=1)

x_num.fillna(x_num.mean(), inplace=True)

std_scl = StandardScaler()

std_scl.fit(x_num)

x_num_std = std_scl.transform(x_num)
x_num_std
```

```python
x_num_std = pd.DataFrame(x_num_std, columns=std_scl.feature_names_in_)
x_num_std
```

```python
x_final = pd.concat([x_num_std, x_cat_ohe], axis=1)
x_final
```

```python
y = df['Species']
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log = LogisticRegression()
svc = SVC()

log.fit(x_final, y)
```

```python
log.score(x_final, y)
```

```python
svc.fit(x_final, y)
svc.score(x_final, y)
```
